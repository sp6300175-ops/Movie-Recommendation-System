# **Problem Statement: Movie Recommendation System 🎬**

With thousands of movies released across various genres, users often struggle to find a movie that suits their preferences. The goal of this project is to create a Movie Recommendation System that suggests movies based on a user's preferred genre and desired number of recommendations. The system should store a collection of movies categorized by genres and use basic Python data structures (lists and dictionaries) to provide recommendations. Additionally, users should be able to customize their recommendations, add new movies, and search for specific movies.

📌 Problem Statement:
Design a Python-based Movie Recommendation System that:

* ✅ Allows users to select a movie genre (Action, Comedy, Drama, Sci-Fi, Horror, etc.).
* ✅ Lets users decide how many recommendations they want.
* ✅ Provides random movie recommendations from the selected genre.
* ✅ Enables users to add new movies to the stored movie list.
* ✅ Stores and retrieves the movie collection using dictionaries.
* ✅ Allows users to search for a movie by title.

📌 Project Objectives:
* 🔹 Genre-Based Movie Recommendations – Suggest movies based on user-selected genre.
* 🔹 Customizable Number of Recommendations – Users can choose how many movie suggestions they receive.
* 🔹 Randomized Suggestions – Ensure variety by picking random movies from the stored collection.
* 🔹 Add New Movies – Users can expand the movie list by adding their own favorites.
* 🔹 Dictionary-Based Storage – Store and update the movie collection in a Python dictionary.
* 🔹 Search Functionality – Allow users to find a movie by title.

📌 Example User Flow:
* Step 1: User runs the program.
* Step 2: System asks the user to select a genre (e.g., "Comedy").
* Step 3: System asks how many movie recommendations they want (e.g., "5").
* Step 4: System displays 5 random movies from the Comedy genre.
* Step 5: User chooses to add a new movie (e.g., "Superbad" under Comedy).
* Step 6: System updates the movie list and saves it.
* Step 7: User searches for a movie by title to check if it exists in the stored collection.

```python
import random

movies={
"comedy":["The Hangover", "Superbad", "Bridesmaids", "Dumb and Dumber", "Groundhog Day",
        "Anchorman", "The Grand Budapest Hotel", "Step Brothers", "Ferris Bueller's Day Off", "Airplane!",
        "Monty Python and the Holy Grail", "21 Jump Street", "The Big Lebowski", "Some Like It Hot", "Mean Girls",
        "Home Alone", "Mrs. Doubtfire", "Borat", "Shaun of the Dead", "Knives Out",
        "Tropic Thunder", "Napoleon Dynamite", "The 40-Year-Old Virgin", "Ghostbusters",
        "Coming to America", "Elf", "Caddyshack", "Office Space", "What We Do in the Shadows", "Pitch Perfect"],

"action":["Die Hard", "Mad Max: Fury Road", "The Dark Knight", "John Wick", "Gladiator",
        "Mission: Impossible – Fallout", "The Matrix", "Terminator 2", "Kill Bill", "Raiders of the Lost Ark",
        "Top Gun: Maverick", "The Bourne Identity", "300", "Lethal Weapon", "Speed",
        "Casino Royale", "The Raid", "Fury", "Extraction", "Crouching Tiger Hidden Dragon",
        "Heat", "Point Break", "The Equalizer", "Atomic Blonde", "Bad Boys",
        "Face/Off", "RoboCop", "Total Recall", "Taken", "Nobody"],

"drama":["The Shawshank Redemption", "Forrest Gump", "Schindler's List", "The Godfather", "12 Years a Slave",
       "A Beautiful Mind", "The Pursuit of Happyness", "Good Will Hunting", "The Green Mile", "Titanic",
       "Moonlight", "Manchester by the Sea", "Marriage Story", "Parasite", "American History X",
       "The Pianist", "Requiem for a Dream", "There Will Be Blood", "One Flew Over the Cuckoo's Nest", "Oppenheimer",
       "The Revenant", "Whiplash", "Dallas Buyers Club", "Room", "Spotlight",
       "The King's Speech", "Brokeback Mountain", "Atonement", "The Imitation Game", "Bohemian Rhapsody"],

"sci-fi":["Interstellar", "Blade Runner 2049", "Inception", "The Matrix", "Star Wars",
        "Arrival", "Alien", "2001: A Space Odyssey", "The Terminator", "E.T.",
        "Dune", "Ex Machina", "District 9", "Gravity", "The Martian",
        "Edge of Tomorrow", "Looper", "Annihilation", "Her", "WALL·E",
        "Avatar", "Minority Report", "Children of Men", "Moon", "Elysium",
        "Oblivion", "Sunshine", "Prometheus", "Tenet", "Everything Everywhere All at Once"],

"horror":["The Exorcist", "Hereditary", "Get Out", "The Shining", "A Nightmare on Elm Street",
          "The Conjuring", "Halloween", "It", "The Ring", "Psycho",
          "The Babadook", "Midsommar", "The Texas Chain Saw Massacre", "Scream", "A Quiet Place",
          "The Witch", "Paranormal Activity", "Sinister", "It Follows", "Barbarian",
          "Insidious", "The Descent", "28 Days Later", "The Omen", "Saw",
          "Candyman", "Us", "The Orphanage", "Smile", "Pearl"],

"thriller":["Se7en", "The Silence of the Lambs", "Gone Girl", "Shutter Island", "Zodiac",
            "Prisoners", "No Country for Old Men", "Nightcrawler", "Black Swan", "Memento",
            "Sicario", "The Departed", "Fight Club", "The Sixth Sense", "Misery",
            "The Girl with the Dragon Tattoo", "Oldboy", "Cape Fear", "Don't Breathe", "The Invisible Man",
            "Split", "Uncut Gems", "Wind River", "The Talented Mr. Ripley", "Panic Room",
            "The Game", "Identity", "Primal Fear", "The Machinist", "10 Cloverfield Lane"]

}

#menu
print("Welcome to the Movie Recommendation System!\n")
print("Please select an option:")
print("1. Get Movie Recommendations")
print("2. Add a New Movie")
print("3. Search for a Movie")
print("4. Exit the Program")


def get_recommendations():
  print("Please select a genre:")
  print("1. Comedy")
  print("2. Action")
  print("3. Drama")
  print("4. Sci-Fi")
  print("5. Horror")
  print("6. Thriller")
  genre=input("Enter your choice:").lower()
  num_recommendations=int(input("Enter the number of recommendations you want:"))
  if genre in movies:
    movie_list=movies[genre]
    if num_recommendations<=len(movie_list):
      random_movies=random.sample(movie_list,num_recommendations)
      print("Here are your recommendations:")
      for movie in random_movies:
        print(movie)
    else:
        print(f"Sorry only {len(movie_list)} movies are available.")
  else:
    print("Invalid genre. Please try again!")


def add_movie():
  print("Please select a genre:")
  print("1. Comedy")
  print("2. Action")
  print("3. Drama")
  print("4. Sci-Fi")
  print("5. Horror")
  print("6. Thriller")
  genre=input("Enter genre of the movie:").lower()
  if genre in movies:
    movie_list=movies[genre]
    new_movie=input("Enter the name of the new movie:")
    if new_movie in movie_list:
      print("Movie already exists in the genre.")
      return
    else:
      movie_list.append(new_movie)
      print("Movie added successfully!")
  else:
    print("Invalid genre. Please try again!")


def search_movie():
  search_term=input("Enter the name of the movie you want to search for:").lower()
  found=False
  for genre,movie_list in movies.items():
    if search_term in [movie.lower() for movie in movie_list]:
      print(f"Movie found in the {genre} genre!")
      found=True
      break
  if not found:
    print("Movie not found in any genre.")


def exit():
  print("Thank you for using the Movie Recommendation System!")

while True:
  choice=int(input("Enter 1/2/3/4 to choose an option:"))

  if choice==1:
    get_recommendations()
  elif choice==2:
    add_movie()
  elif choice==3:
    search_movie()
  elif choice==4:
    exit()
    break
  else:
    print("Invalid choice. Please try again!")
```
